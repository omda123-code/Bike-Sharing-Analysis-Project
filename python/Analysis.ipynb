```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
os.makedirs('plots',exist_ok=True)
```

```python
sns.set_style("whitegrid")
```

```python
# 1. Total rides per day
daily_total = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Bike/outputs/daily_total_rides.csv')
daily_total['dteday'] = pd.to_datetime(daily_total['dteday'])
weekly_total = daily_total.resample('W', on='dteday').sum()

plt.figure(figsize=(12,6))
plt.plot(weekly_total.index, weekly_total['cnt'], marker='o')
plt.title('Weekly Total Rides')
plt.xlabel('Week')
plt.ylabel('Total Rides')
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig('plots/Weekly_Total_Rides.png')
plt.show()
```

```python
daily_total['cnt_smooth'] = daily_total['cnt'].rolling(7).mean()

plt.figure(figsize=(12,6))
plt.plot(daily_total['dteday'], daily_total['cnt_smooth'], label='7-Day Rolling Average')
plt.title('Total Rides per Day (Smoothed)')
plt.xlabel('Date')
plt.ylabel('Total Rides')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig('plots/Total_Rides_per_Day.png')
plt.show()
```

```python
# 2. Bike usage by hour
hourly_usage = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Bike/outputs/hourly_usage.csv')
plt.figure(figsize=(10,5))
sns.barplot(x='hr', y='total_rentals', data=hourly_usage,hue='hr', palette='viridis')
plt.title('Bike Usage by Hour of the Day')
plt.xlabel('Hour')
plt.ylabel('Total Rentals')
plt.tight_layout()
plt.savefig('plots/hourly_usage.png')
plt.show()
```

```python
# 12. Daily trend of casual vs registered users
daily_user_trend = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Bike/outputs/daily_user_trend.csv')
daily_user_trend['dteday'] = pd.to_datetime(daily_user_trend['dteday'])
weekly_trend = daily_user_trend.resample('W', on='dteday').sum()

plt.figure(figsize=(12,6))
plt.plot(weekly_trend.index, weekly_trend['total_casual'], label='Casual', marker='o')
plt.plot(weekly_trend.index, weekly_trend['total_registered'], label='Registered', marker='o')
plt.title('Weekly Trend of Casual vs Registered Users')
plt.xlabel('Week')
plt.ylabel('Total Users')
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.savefig('plots/Weekly_Trend_of_Casual_vs_Registered_Users.png')
plt.show()
```

```python
# 13. Season-wise ratio of casual vs registered
season_user_ratio = pd.read_csv('C:/Users/Ismail/Desktop/Projects for GitHup/Sql/Bike/outputs/season_user_ratio.csv')
plt.figure(figsize=(8,5))
season_user_ratio.set_index('season')[['total_casual','total_registered']].plot(kind='bar', stacked=True)
plt.title('Proportion of Casual vs Registered Users per Season')
plt.xlabel('Season')
plt.ylabel('Total Users')
plt.tight_layout()
plt.savefig('plots/season_user_ratio.png')
plt.show()
```
